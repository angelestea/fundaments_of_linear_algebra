```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

```python
v1 = np.array([1,1])
v2 = np.array([-1,-1])

fig = plt.figure()
ax=fig.add_subplot(111, projection='3d')
```

```python
v1 = np.array([1,1])
v2 = np.array([-1,-1])

fig = plt.figure()
ax=fig.add_subplot(111, projection='3d')

for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a +v2[0]*b, v1[1]*a+v2[1]*b,
                    v1[2]*a+v2[2]*b,
                   marker = '.',
                   color='orange')
        
ax.set_xlabel('Eje x')
ax.set_ylabel('Eje y')
ax.set_zlabel('Eje z')

plt.show()
```
